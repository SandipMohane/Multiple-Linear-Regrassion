# calculating VIF 

```python
df.columns
```

```python
r_hp = smf.ols('HP~VOL+SP+WT',data=df).fit().rsquared
v_hp = 1/(1-r_hp)

r_vol = smf.ols('VOL~SP+HP+WT',data=df).fit().rsquared
v_vol = 1/(1-r_vol)

r_sp = smf.ols('SP~VOL+WT+WT',data=df).fit().rsquared
v_sp = 1/(1-r_sp)

r_wt = smf.ols('WT~VOL+SP+SP',data=df).fit().rsquared
v_wt = 1/(1-r_wt)
```

```python
vif = pd.DataFrame({ 
       'feature_name': ['HP','VOL','SP','WT'],
       'vif': [v_hp,v_vol,v_sp,v_wt]

})
```

```python
vif
```

```python
df.drop(columns=['VOL'],inplace=True)# drop the column whos vif value is greater 
```

```python
df
```

##  Creating Final Model

```python
fmodel = smf.ols('MPG~HP+SP+WT',data=df).fit()
```
